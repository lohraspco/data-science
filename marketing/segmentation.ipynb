```python
import os
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
os.environ['OMP_NUM_THREADS'] = '2'

# Example: Simulated Marketing Data for General Motors
# Features: [Annual Income, Spending Score]
data = {
    'Annual_Income': np.random.normal(60000, 15000, 500),
    'Spending_Score': np.random.normal(50, 20, 500)
}
df = pd.DataFrame(data)

# Convert to numpy array
X = df.values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Fit GMM model
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm.fit(X_scaled)

# Predict cluster labels
labels = gmm.predict(X_scaled)
df['Cluster'] = labels

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(df['Annual_Income'], df['Spending_Score'], c=labels, cmap='viridis', s=30, alpha=0.6)
plt.title('Customer Segmentation for General Motors')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.colorbar(label='Cluster')
plt.show()
```
